### Load Libraries and Import Data

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline

# Plot style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 12
```

```python
# Load dataset
# Download from: https://www.kaggle.com/datasets/valakhorasani/gym-members-exercise-dataset
# Then place the CSV in the same folder as this notebook

df = pd.read_csv('gym_members_exercise_tracking.csv')

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
# First look at the data
df.head(10)
```

```python
# Column names and data types
df.info()
```

```python
# Basic statistics
df.describe().round(2)
```

## 🧹 Step 3: Process

In this step we will:
- Check for missing values
- Check for duplicate rows
- Verify data types
- Add useful new columns
- Clean column names


```python
# ── Check for missing values ──────────────────────────────────────────────────
print("=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "✅ No missing values found!")
```

```python
# ── Check for duplicate rows ──────────────────────────────────────────────────
duplicates = df.duplicated().sum()
print(f"=== Duplicate Rows ===")
print(f"{'✅ No duplicates found!' if duplicates == 0 else f'⚠️ {duplicates} duplicate rows found.'}")
```

```python
# ── Clean column names (remove spaces, lowercase) ─────────────────────────────
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

print("=== Cleaned Column Names ===")
print(df.columns.tolist())
```

```python
# ── Map Experience Level to readable labels ───────────────────────────────────
experience_map = {1: 'Beginner', 2: 'Intermediate', 3: 'Expert'}
df['experience_label'] = df['experience_level'].map(experience_map)

# ── Add BMI Category column ────────────────────────────────────────────────────
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['bmi_category'] = df['bmi'].apply(bmi_category)

# ── Add Calories per Hour column ──────────────────────────────────────────────
df['calories_per_hour'] = (df['calories_burned'] / df['session_duration_hours']).round(1)

print("✅ New columns added: experience_label, bmi_category, calories_per_hour")
df[['experience_level', 'experience_label', 'bmi', 'bmi_category', 'calories_per_hour']].head()
```

```python
# ── Final check on cleaned dataset ────────────────────────────────────────────
print(f"✅ Clean dataset ready: {df.shape[0]} rows × {df.shape[1]} columns")
df.head()
```

## 🔍 Step 4: Analyze

Now we answer our business questions using data.

### Questions:
1. What workout types are most popular?
2. How does workout frequency relate to calories burned?
3. Is there a gender difference in fitness metrics?
4. What is the BMI distribution of members?
5. How does experience level affect performance?


### Q1. What workout types are most popular?

```python
workout_counts = df['workout_type'].value_counts().reset_index()
workout_counts.columns = ['Workout Type', 'Count']
workout_counts['Percentage'] = (workout_counts['Count'] / len(df) * 100).round(1)
print(workout_counts)
```

```python
# Summary stats by workout type
df.groupby('workout_type')[['calories_burned', 'session_duration_hours', 'avg_bpm']].mean().round(2)
```

### Q2. How does workout frequency relate to calories burned?

```python
freq_calories = df.groupby('workout_frequency_days/week')['calories_burned'].agg(['mean', 'median', 'count']).round(2)
freq_calories.columns = ['Mean Calories', 'Median Calories', 'Count']
print(freq_calories)
```

### Q3. Gender differences in fitness metrics

```python
gender_stats = df.groupby('gender')[['calories_burned', 'fat_percentage', 'bmi', 'avg_bpm', 'session_duration_hours']].mean().round(2)
print(gender_stats)
```

### Q4. BMI distribution

```python
bmi_dist = df['bmi_category'].value_counts().reset_index()
bmi_dist.columns = ['BMI Category', 'Count']
bmi_dist['Percentage'] = (bmi_dist['Count'] / len(df) * 100).round(1)
print(bmi_dist)
```

### Q5. Experience level vs. performance

```python
exp_stats = df.groupby('experience_label')[['calories_burned', 'session_duration_hours', 'fat_percentage', 'water_intake_liters']].mean().round(2)
# Order correctly
exp_stats = exp_stats.reindex(['Beginner', 'Intermediate', 'Expert'])
print(exp_stats)
```

## 📊 Step 5: Share

We will now create visualizations to communicate our findings clearly.


### Chart 1: Most Popular Workout Types

```python
fig, ax = plt.subplots(figsize=(9, 5))

colors = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
workout_counts_plot = df['workout_type'].value_counts()

bars = ax.bar(workout_counts_plot.index, workout_counts_plot.values, color=colors, edgecolor='white', linewidth=1.2)

for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 4,
            f'{bar.get_height()}', ha='center', va='bottom', fontweight='bold', fontsize=11)

ax.set_title('Most Popular Workout Types Among Gym Members', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Workout Type', fontsize=12)
ax.set_ylabel('Number of Members', fontsize=12)
ax.set_ylim(0, workout_counts_plot.max() + 50)
sns.despine()
plt.tight_layout()
plt.show()
```

### Chart 2: Workout Frequency vs. Average Calories Burned

```python
freq_avg = df.groupby('workout_frequency_days/week')['calories_burned'].mean().reset_index()

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(freq_avg['workout_frequency_days/week'], freq_avg['calories_burned'],
        marker='o', color='#4C72B0', linewidth=2.5, markersize=9, markerfacecolor='white', markeredgewidth=2.5)

for _, row in freq_avg.iterrows():
    ax.annotate(f"{row['calories_burned']:.0f}",
                xy=(row['workout_frequency_days/week'], row['calories_burned']),
                xytext=(0, 12), textcoords='offset points', ha='center', fontsize=10, fontweight='bold')

ax.set_title('Workout Frequency vs. Average Calories Burned', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Workout Frequency (days/week)', fontsize=12)
ax.set_ylabel('Average Calories Burned', fontsize=12)
ax.set_xticks(freq_avg['workout_frequency_days/week'])
sns.despine()
plt.tight_layout()
plt.show()
```

### Chart 3: Gender Comparison — Key Fitness Metrics

```python
metrics = ['calories_burned', 'fat_percentage', 'session_duration_hours']
labels = ['Calories Burned', 'Fat Percentage (%)', 'Session Duration (hrs)']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

colors = {'Male': '#4C72B0', 'Female': '#DD8452'}

for ax, metric, label in zip(axes, metrics, labels):
    data_m = df[df['gender'] == 'Male'][metric]
    data_f = df[df['gender'] == 'Female'][metric]

    bp = ax.boxplot([data_m, data_f], labels=['Male', 'Female'], patch_artist=True,
                    medianprops=dict(color='black', linewidth=2))

    for patch, color in zip(bp['boxes'], ['#4C72B0', '#DD8452']):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)

    ax.set_title(label, fontsize=12, fontweight='bold')
    ax.set_ylabel(label, fontsize=10)

fig.suptitle('Gender Comparison Across Key Fitness Metrics', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### Chart 4: BMI Category Distribution

```python
bmi_order = ['Underweight', 'Normal', 'Overweight', 'Obese']
bmi_counts = df['bmi_category'].value_counts().reindex(bmi_order, fill_value=0)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Bar chart
colors_bmi = ['#55A868', '#4C72B0', '#DD8452', '#C44E52']
axes[0].bar(bmi_counts.index, bmi_counts.values, color=colors_bmi, edgecolor='white')
for i, (label, val) in enumerate(zip(bmi_counts.index, bmi_counts.values)):
    axes[0].text(i, val + 3, str(val), ha='center', fontweight='bold')
axes[0].set_title('BMI Category Counts', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Number of Members')
sns.despine(ax=axes[0])

# Pie chart
axes[1].pie(bmi_counts.values, labels=bmi_counts.index, autopct='%1.1f%%',
            colors=colors_bmi, startangle=90,
            wedgeprops=dict(edgecolor='white', linewidth=1.5))
axes[1].set_title('BMI Category Share (%)', fontsize=13, fontweight='bold')

fig.suptitle('BMI Distribution of Gym Members', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Chart 5: Experience Level vs. Calories Burned & Fat Percentage

```python
exp_order = ['Beginner', 'Intermediate', 'Expert']
exp_data = df[df['experience_label'].isin(exp_order)]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
palette = {'Beginner': '#55A868', 'Intermediate': '#4C72B0', 'Expert': '#C44E52'}

# Calories
means_cal = exp_data.groupby('experience_label')['calories_burned'].mean().reindex(exp_order)
axes[0].bar(means_cal.index, means_cal.values,
            color=[palette[l] for l in means_cal.index], edgecolor='white')
for i, val in enumerate(means_cal.values):
    axes[0].text(i, val + 5, f'{val:.0f}', ha='center', fontweight='bold')
axes[0].set_title('Avg Calories Burned by Experience', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Average Calories Burned')
sns.despine(ax=axes[0])

# Fat %
means_fat = exp_data.groupby('experience_label')['fat_percentage'].mean().reindex(exp_order)
axes[1].bar(means_fat.index, means_fat.values,
            color=[palette[l] for l in means_fat.index], edgecolor='white')
for i, val in enumerate(means_fat.values):
    axes[1].text(i, val + 0.3, f'{val:.1f}%', ha='center', fontweight='bold')
axes[1].set_title('Avg Fat Percentage by Experience', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Average Fat Percentage (%)')
sns.despine(ax=axes[1])

fig.suptitle('Fitness Metrics by Experience Level', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Chart 6: Correlation Heatmap

```python
numeric_cols = ['age', 'weight_kg', 'session_duration_hours', 'calories_burned',
                'fat_percentage', 'water_intake_liters', 'workout_frequency_days/week',
                'bmi', 'avg_bpm', 'resting_bpm']

corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(11, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, ax=ax,
            annot_kws={'size': 9})

ax.set_title('Correlation Heatmap of Fitness Variables', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.show()
```

### Chart 7: Calories Burned vs. Session Duration (by Workout Type)

```python
fig, ax = plt.subplots(figsize=(10, 6))

palette_wt = {'Cardio': '#4C72B0', 'Strength': '#DD8452', 'Yoga': '#55A868', 'HIIT': '#C44E52'}

for wtype, group in df.groupby('workout_type'):
    ax.scatter(group['session_duration_hours'], group['calories_burned'],
               label=wtype, color=palette_wt.get(wtype, 'gray'),
               alpha=0.5, s=40, edgecolors='none')

ax.set_title('Calories Burned vs. Session Duration by Workout Type', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Session Duration (hours)', fontsize=12)
ax.set_ylabel('Calories Burned', fontsize=12)
ax.legend(title='Workout Type', frameon=True)
sns.despine()
plt.tight_layout()
plt.show()
```

## 💡 Step 6: Act

### Summary of Findings

Based on our analysis of 973 gym members' exercise data, here are the key findings:

| # | Finding |
|---|---|
| 1 | **All 4 workout types are equally popular** — Cardio, Strength, Yoga, and HIIT each account for roughly 25% of sessions. |
| 2 | **Higher workout frequency = more calories burned** — Members who work out 5 days/week burn significantly more calories than those who work out 2 days/week. |
| 3 | **Male members burn slightly more calories** on average, but female members have lower fat percentages in higher experience groups. |
| 4 | **Most members (≈60%) are in the Normal or Overweight BMI range** — suggesting room for targeted health improvement programs. |
| 5 | **Expert-level members burn far more calories** and have much lower body fat than beginners — showing the clear benefit of long-term gym commitment. |
| 6 | **Strong positive correlation** between session duration, workout frequency, and calories burned. Water intake also rises with fitness level. |

---

### Recommendations for FitTrack App

**1. 📅 Build a Workout Streak Feature**  
Since higher frequency leads to more calories burned, a streak/habit tracker can motivate users to maintain consistency.

**2. 🎯 Personalize by Experience Level**  
Beginners and experts have very different profiles. Offer beginner-friendly guided plans vs. advanced performance tracking for experts.

**3. 🍎 Add BMI & Body Composition Tracking**  
Most users are in the Normal/Overweight range. A built-in BMI tracker with personalized feedback would add meaningful value.

**4. 💧 Promote Hydration Reminders**  
More active users drink more water. Sending smart water reminders based on workout intensity can improve user health outcomes.

**5. 🏆 Gamify Calorie Goals by Workout Type**  
Since HIIT and Strength training tend to burn the most calories, nudge less-active users toward these formats.
